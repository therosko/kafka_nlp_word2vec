```python
#    Spark
from pyspark import SparkContext
#    Spark Streaming
from pyspark.streaming import StreamingContext
#    Kafka
from pyspark.streaming.kafka import KafkaUtils
#    json parsing
import json
import pprint
```

```python
sc = SparkContext(appName="PythonSparkStreamingKafka_RM_01")
sc.setLogLevel("WARN")
```

```python
ssc = StreamingContext(sc, 30)
```

```python
kafkaStream = KafkaUtils.createStream(ssc, 'localhost:2181', 'spark-streaming', {'elon':1})
```

```python
# Kafka stores things as name value pairs, therefore we need the value which in our case is tweets in json format
# .filter(lambda x: len(x)>0) is used to filter out non-json messages
jsons_dstream = kafkaStream.map(lambda x: json.loads(x[1])).filter(lambda x: len(x)>0)

# Extract the text and the created metadata
csv_dstream = jsons_dstream.map(lambda j_msg: '%s \n\n'%(j_msg.get('text')))

#csv_dstream.pprint(100)
csv_dstream.saveAsTextFiles("data/elon","txt") #filepath and prefix, suffix=None

ssc.start()
ssc.awaitTermination() #It just waits for the termination signal from user. When it receives signal from user (i.e CTRL+C or SIGTERM) then it streaming context will be stopped. It is kind of shutdownhook in java.
```
